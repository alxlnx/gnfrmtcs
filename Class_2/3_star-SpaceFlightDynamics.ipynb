### Элементы динамики космического полета

Во многих случаях можно получить достаточно хорошее представление о движении какого-либо небесного тела (по крайней мере в течение небольшого промежутка времени), если учесть воздействие на него лишь одного, наиболее сильно притягивающего его тела и пренебречь влиянием всех других небесных тел. Так, например, обстоит дело при изучении движения Земли вокруг Солнца или близкого искусственного спутника вокруг Земли.

Пусть нас интересует движение тела $m_1$ в гравитационном поле, созданном другим телом $m_2$. Это движение будем рассматривать в системе отсчета с началом в барицентре (центре масс) тела $m_2$ и с осями, постоянно ориентированными в пространстве. Если размеры тела $m_1$ малы по сравнению с расстояниями его точек до точек тела $m_2$, то мы получим достаточно хорошее представление о движении любой его точки, если изучим движение той материальной точки, которая образуется, если всю массу тела $m_1$ сосредоточить в его барицентре. В результате такого сосредоточения массы траектория барицентра тела $m_1$ практически не изменится. Когда говорят о траектории некоторого тела $m_1$ и о его скорости, имеют в виду траекторию и скорость материальной точки. Что касается тела $m_2$, то будем считать, что оно является шаром со сферическим распределением плотности (если во всех точках, равноудаленных от центра шара, плотности равны, то говорят, что шар имеет сферическое распределение плотности). Поэтому сила, действующая на материальную точку $m_1$, не изменится, если мы будем считать всю массу тела $m_2$ сосредоточенной в его барицентре.

Таким образом, мы приходим к следующей задаче: изучить движение точечной массы, происходящее в гравитационном поле, созданном некоторым телом $m_2$, если это тело является шаром со сферическим распределением плотности.

При сделанном выше предположении о теле $m_2$ эта задача равносильна, очевидно, такой задаче: изучить движение материальной точки в гравитационном поле, созданном другой материальной точкой. При этом движение будем рассматривать относительно точки $m_2$, то есть в системе отсчета с началом в точке $m_2$ и с осями, сохраняющими ориентацию в пространстве. Эту задачу обычно называют задачей двух тел (точнее было бы назвать ее «задачей двух материальных точек»).


```python
%pylab inline

fig, ax = plt.subplots(figsize=(7, 8))

vec_x, vec_y = [0, 0, 0, 4, 4], [0, 0, 0, 4, 4]
vec_u, vec_v = [1, 6, 4, -3, 2], [7, 2, 4, 3, -2]
qkeys_x, qkeys_y = [0.57, 3.5, 2, 2.5, 5], [4, 1.16, 2, 5.5, 3]
qkeys = ["$\\vec{R_1}$", "$\\vec{R_2}$", "$\\vec{R_C}$", "$\\vec{r_1}$", "$\\vec{r_2}$"]

for x, y, u, v, q_x, q_y, q_key in zip(vec_x, vec_y, vec_u, vec_v, qkeys_x, qkeys_y, qkeys):
    q = ax.quiver(x, y, u, v, angles='xy', scale_units='xy', scale=1)
    ax.quiverkey(q, q_x, q_y, 0, q_key, coordinates='data', 
                 labelpos='W' if q_key == "$\\vec{R_1}$" else 'N', 
                 fontproperties={'size': 15})

x, y = [0, 1, 4, 6], [0, 7, 4, 2]
xytext_shift = [-0.45, 0.15, 0.15, 0.15]
ann = ["O", "$m_1$", "C", "$m_2$"]

ax.scatter(x, y, s=100)

for i, txt in enumerate(ann):
    ax.annotate(txt, (x[i], y[i]), xytext=(x[i] + xytext_shift[i], y[i] + xytext_shift[i]), size=15)

plt.xlim(-1, 7)
plt.ylim(-1, 8)
plt.axis('off');
```

Рассмотрим систему, состоящую из двух тел – материальных точек (далее – точек) с массами $m_1$ и $m_2$, положение которых в инерциальном пространстве задается соответственно векторами $\vec{R_1}$ и $\vec{R_2}$ относительно некоторой точки $O$.

Согласно закону всемирного тяготения на точку массой $m_1$ со стороны точки массой $m_2$ действует сила 

$$\large \vec{F_1}=\gamma m_1 m_2 \frac{1}{r^2} \frac{\vec{r}}{r},$$

где $\gamma$ – универсальная гравитационная постоянная, $\vec{r}=\vec{R_2}-\vec{R_1}$ – радиус-вектор точки с массой $m_2$ относительно точки с массой $m_1$, $r=|\vec{r}|$. 

Поле, обладающее указанной зависимостью силы от расстояния, носит название «центральное ньютоново поле». Уравнения движения этих двух материальных точек имеют вид 

$$\large m_1 \frac{d^2\vec{R_1}}{dt^2} = \vec{F_1},\ m_2 \frac{d^2\vec{R_2}}{dt^2} = \vec{F_2} \tag{1}.$$

Складывая эти два уравнения с учетом соотношения между действующими на точки силами $\vec{F_1} = -\vec{F_2}$, получаем 

$$\large\frac{d^2(m_1\vec{R_1} + m_2\vec{R_2})}{dt^2} = 0. \tag{*}$$

Введем барицентр (центр масс) системы этих двух точек. Его положение определяется радиус-вектором $\vec{R_C}$ удовлетворяющим соотношению 

$$\large m_1\vec{R_1} + m_2\vec{R_2}=(m_1+m_2)\vec{R_C} \tag{2}.$$ 

Тогда, подставив $(2)$ в $(*)$ получаем уравнение, описывающее движение барицентра системы: $\frac{d^2\vec{R_C}}{dt^2} = 0$, где произведено деление на положительную величину $m_1+m_2$. Дважды интегрируя это уравнение, получаем два векторных первых интеграла

$$\large \frac{d\vec{R_C}}{dt}\equiv\vec{V_C}=\vec{C_1},\ \vec{R_C}=\vec{V_C}t+\vec{C_2}.$$ 

Здесь $\vec{C_1}, \vec{C_2}$ – постоянные первых интегралов и введено обозначение $\vec{V_C}$ для скорости центра масс системы. 

Первым интегралом называется функция, зависящая от координат, скоростей и времени, остающаяся постоянной в силу уравнений движения на любых движениях системы. Полученные первые интегралы отражают вполне очевидный факт, что центр масс системы, не подверженной действию внешних сил, движется прямолинейно и равномерно в инерциальном пространстве.

Перенесем начало координат в центр масс системы. Для этого подставим выражения $\vec{R_j}=\vec{R_C}+\vec{r_j} (j=1, 2)$ в $(1)$ и $(2)$:

$$m_1(\vec{R_C}+\vec{r_1}) + m_2(\vec{R_C}+\vec{r_2})=(m_1+m_2)\vec{R_C} \implies m_1\vec{r_1}+m_2\vec{r_2}=0,$$

$$m_j\frac{d^2(\vec{R_C}+\vec{r_j})}{dt^2}=\vec{F_j},\ \frac{d^2\vec{R_C}}{dt^2} = 0 \implies m_j\frac{d^2\vec{r_j}}{dt^2}=\vec{F_j},$$

откуда получаем $m_1\vec{r_1} + m_2\vec{r_2} = 0$ и $m_1\frac{d^2\vec{r_1}}{dt^2}=\vec{F_1}$, $m_2\frac{d^2\vec{r_2}}{dt^2}=\vec{F_2}$. Здесь $\vec{r_1}$ и $\vec{r_2}$ – радиус-векторы точек относительно их общего центра масс. 

Выражая $\vec{r_2} = -\frac{m_1\vec{r_1}}{m_2}$ и подставляя в $\vec{r}=\vec{r_2} - \vec{r_1}$:

$$\vec{r}=-\vec{r_1}\left(\frac{m_1}{m_2}+1\right) = -\vec{r_1}\frac{1}{m_2}(m_1+m_2) \implies \vec{F_1}=-\gamma m_1 m_2 \frac{\vec{r_1}}{r^3} \frac{1}{m_2} (m_1 + m_2) = -\gamma m_1 (m_1 + m_2) \frac{\vec{r_1}}{r^3}, $$

тогда формула для гравитационной силы, действующей на точку с массой $m_1$, принимает вид $\vec{F_1} = -\gamma m_1 (m_1 + m_2) \frac{\vec{r_1}}{r^3}$. 

Введем гравитационный параметр системы $\mu = \gamma(m_1+m_2)$, тогда

$$ \vec{F_1} = -\gamma m_1 (m_1 + m_2) \frac{\vec{r_1}}{r^3} =  -\mu m_1 \frac{\vec{r_1}}{r^3}, \ \vec{F_1} = m_1 \frac{d^2\vec{r_1}}{dt^2}$$

то есть $\frac{d^2\vec{r_1}}{dt^2} = -\mu\frac{\vec{r_1}}{r^3}$. Аналогично: $\frac{d^2\vec{r_2}}{dt^2} = -\mu\frac{\vec{r_2}}{r^3}$. Вспоминая, что $\vec{r}=\vec{r_2} - \vec{r_1}$:

$$ \frac{d^2\vec{r_1}}{dt^2} - \frac{d^2\vec{r_2}}{dt^2} = -\mu\left(\frac{\vec{r_2}}{r^3} - \frac{\vec{r_1}}{r^3}\right) \implies$$

$$\large \frac{d^2\vec{r}}{dt^2}=-\mu\frac{\vec{r}}{r^3} \tag{3}.$$

Последние три уравнения показывают, что движение каждой точки вокруг общего центра масс, как и движение одной относительно другой, подобны. Последнее же уравнение можно интерпретировать как уравнение, описывающее движение точки в поле притягивающего центра.

Вводя проекции $x$, $y$, $z$ вектора $\vec{r}$ на оси инерциальной системы координат, можно переписать последнее уравнение в координатном виде $\frac{d^2x}{dt^2} = -\mu\frac{x}{r^3}$, $\frac{d^2y}{dt^2} = -\mu\frac{y}{r^3}$, $\frac{d^2z}{dt^2}=-\mu\frac{z}{r^3}$.

Умножим левую и правую части уравнения $(3)$ векторно на $\vec{r}$: $\vec{r}\times\frac{d^2\vec{r}}{dt^2}=0$. Продифференцируем выражение $\frac{d}{dt}\left(\vec{r}\times\frac{d\vec{r}}{dt}\right)$:

$$\frac{d}{dt}\left(\vec{r}\times\frac{d\vec{r}}{dt}\right)=\frac{d\vec{r}}{dt}\times\frac{d\vec{r}}{dt}+\vec{r}\times\frac{d^2\vec{r}}{dt^2}=\vec{r}\times\frac{d^2\vec{r}}{dt^2}.$$

Следовательно, предыдущее уравнение приобретает вид $\frac{d}{dt}\left(\vec{r}\times\frac{d\vec{r}}{dt}\right)=0$. Его интегрирование дает **интеграл площадей** $\vec{r}\times\frac{d\vec{r}}{dt}=\vec{c}$. Который можно представить в виде 

$$\large\vec{r}\times\vec{V}=\vec{c} \tag{4}.$$ 

Постоянная интегрирования – вектор $\vec{c}$ – по сути, это вектор кинетического момента, отнесенного к массе системы. Интеграл площадей означает, что секториальная скорость тела относительно притягивающего центра постоянна (секториальная скорость - площадь, которую ометает радиус-вектор тела за единицу времени при движении вдоль кривой).

Если обозначить через $i$, $j$, $k$ орты инерциальной системы координат, центр которой совпадает с притягивающей точкой, то можно записать $\vec{c}=c_1i+c_2j+c_3k$, $\vec{r}=xi+yj+zk$, и интеграл площадей приобретает вид

$$c_1=y\frac{dz}{dt}-z\frac{dy}{dt}, c_2=z\frac{dx}{dt}-x\frac{dz}{dt}, c_3=x\frac{dy}{dt}-y\frac{dx}{dt}.$$

Умножим левую и правые части выражения $(4)$ скалярно на $\vec{r}$:

$$\vec{r}\cdot(\vec{r}\times\vec{V}) = \vec{r}\cdot\vec{c},$$ 

$$\vec{r}\cdot(\vec{r}\times\vec{V}) = r_x(r_y V_z - r_z V_y) + r_y(r_z V_x - r_x V_z) + r_z(r_x V_y - r_y V_x) = 0 \implies \vec{r}\cdot\vec{c} = 0,$$

и запишем скалярное произведение в координатном виде

$$\large c_1x+c_2y+c_3z=0 \tag{5}.$$

Следовательно, координаты $x$, $y$, $z$ движущегося тела удовлетворяют уравнению плоскости, идущей через начало координат, совпадающее с центральным телом, и ортогональной вектору $\vec{c}$. Таким образом, движение тела – плоское.

Теперь умножим правую и левую части уравнения $(3)$ справа на вектор $\vec{c}$: $\frac{d^2\vec{r}}{dt^2}\times\vec{c}=-\mu\frac{\vec{r}}{r^3}\times\vec{c}$. В левой части заменим $\frac{d\vec{r}}{dt}$ на $\vec{V}$ и внесем вектор $\vec{c}$ под знак производной. Воспользуемся выражением для двойного векторного произведения $\vec{a}\times\left(\vec{b}\times\vec{c}\right)=\vec{b}\left(\vec{a}\cdot\vec{c}\right)-\vec{c}(\vec{a}\cdot\vec{b})$ в правой части, подставив выражение для $\vec{c}$: 

$$\frac{d}{dt}\left(\vec{V}\times\vec{c}\right) = -\frac{\mu}{r^3}(\vec{r}\times(\vec{r}\times\vec{V})) = -\frac{\mu}{r^3}(\vec{r}\cdot(\vec{r}\cdot\vec{V}) - \vec{V}(\vec{r}\cdot\vec{r})) = -\mu\frac{1}{r^3}\left[\vec{r}\left(\vec{r}\cdot\frac{d\vec{r}}{dt}\right)-\frac{d\vec{r}}{dt}\left(\vec{r}\cdot\vec{r}\right)\right]=-\mu\frac{1}{r^3}\left[\vec{r}r\cdot\frac{dr}{dt}-\frac{d\vec{r}}{dt}r^2\right].$$ 

После сокращения числителя и знаменателя в правой части на $r$ и «усмотрения» в выражении, оставшемся в квадратных скобках, производной дроби $\frac{d}{dt}\left(\frac{\vec{r}}{r}\right)=\frac{1}{r^2}\left(\frac{d\vec{r}}{dt}r - \vec{r}\frac{dr}{dt}\right)$, получаем уравнение $\frac{d}{dt}\left(\vec{V}\times\vec{c}\right)=\mu\frac{d}{dt}\frac{\vec{r}}{r}$. Интегрируя его, получаем **интеграл Лапласа**:

$$\large \vec{V}\times\vec{c}-\mu\frac{\vec{r}}{r}=\vec{f}.$$ 

Данный вектор направлен вдоль оси симметрии конического сечения и указывает на перицентр – ближайшую к центральному телу точку орбиты.

Легко убедиться, что интеграл Лапласа $\vec{f}$ связан с интегралом площадей $\vec{c}$ соотношением $\vec{c}\vec{f}=0$: 

$$\vec{c}\vec{f}=\left(\vec{r}\times\vec{V}\right)\left(\vec{V}\times\vec{c}-\mu\frac{\vec{r}}{r}\right)=\vec{c}\left(\vec{V}\times\vec{c}\right)-\mu\left(\vec{r}\times\vec{V}\right)\frac{\vec{r}}{r}=0$$

Последнее равенство в этой цепочке справедливо в силу ортогональности векторов в скалярных произведениях.

Вычислим скалярное произведение радиус-вектора $\vec{r}$ точки относительно притягивающего центра и интеграла Лапласа: 

$$\vec{r}\cdot\vec{f}=\vec{r}\cdot\left(\vec{V}\times\vec{c}-\mu\frac{\vec{r}}{r}\right).$$

Согласно свойству смешанного произведения:

$$\vec{r}\cdot\left(\vec{V}\times\vec{c}\right) = \vec{V}\cdot\left(\vec{c}\times\vec{r}\right) = \vec{c}\cdot\left(\vec{r}\times\vec{V}\right) = c^2$$

Таким образом,

$$\vec{r}\cdot\left(\vec{V}\times\vec{c}-\mu\frac{\vec{r}}{r}\right)=c^2-\mu r.$$ 

Зададим вектор Лапласа его проекциями $f_1$, $f_2$, $f_3$ на оси ранее упомянутой инерциальной системы координат. Тогда траектория движения точки определяется этим произведением, записанным через введенные проекции вектора $\vec{f}$, и полученным ранее уравнением плоскости движения:

$$c^2-xf_1-yf_2-zf_3=\mu r,\ \ c_1x+c_2y+c_3z=0.$$

Введем новую систему координат с осями $O_\xi$, $O_\eta$, $O_\zeta$, направив $O_\xi$ по вектору $\vec{f}$, а $O_\zeta$ – по $\vec{c}$. 

```python
%pylab inline

fig, ax = plt.subplots(figsize=(7, 7))

origin = np.zeros(5)

vectors = np.vstack((np.eye(2), [[-0.3, -0.3]], [[0.5, 0]], [[0, 0.5]]))
qkeys = ["$\\xi$", "$\\zeta$", "$\\eta$", "$\\vec{f}$", "$\\vec{c}$"]
qkeys_x, qkeys_y = [0.95, 0, -0.28, 0.4, 0], [0, 0.95, -0.28, 0, 0.4]
colors = ['k', 'k', 'k', 'r', 'r']
labelposes = ['N', 'W', 'S', 'N', 'W']

for o, u, v, q_x, q_y, q_key, c, lp in zip(origin, vectors[:, 0], vectors[:, 1], 
                                           qkeys_x, qkeys_y, qkeys, colors, labelposes):
    q = ax.quiver(o, o, u, v, angles='xy', scale_units='xy', scale=1, color=c)
    ax.quiverkey(q, q_x, q_y, 0, q_key, coordinates='data', fontproperties={'size': 15}, labelpos=lp)

ax.quiverkey(q, 0, 0.05, 0.05, "$O$", coordinates='data', fontproperties={'size': 20}, labelpos='W')
    
plt.xlim(-0.4, 1.1)
plt.ylim(-0.4, 1.1)
plt.axis('off');
```

Это можно сделать, так как $\vec{c}\vec{f}=0$. Тогда выполняются равенства $c_1=c_2=0$, $c_3=c$, $f_1=f$, $f_2=f_3=0$ и предыдущая система приобретает вид $\mu r=c^2-\xi f$, где $r=\sqrt{\xi^2+\eta^2}$. 

Перейдем к полярным координатам (радиус-вектору $\vec{r}$ и истинной аномалии $\upsilon$ – угловой параметр, который определяет положение тела, движущегося по кеплеровской орбите) по стандартным формулам: $\xi=r\cos(\upsilon)$, $\eta=r\sin(\upsilon)$. 
_Комментарий: на самом деле, корректнее было сказать, что переход был совершен в цилиндрическую систему ($r$, $\upsilon$, $z$), где $z=0$._

Тогда $$\mu r=c^2-f r \cos(\upsilon) \implies r=\frac{c^2}{\mu + f \cos(\upsilon)} \ \ или \ \ r=\frac{p}{1+e \cos(\upsilon)},$$

где 

$$\large p=\frac{c^2}{\mu},\ \ e=\frac{f}{\mu}.$$ 

Полученные переменные $p$ и $e$ называют **(фокальный) параметр орбиты** и **эксцентриситет** соответственно. 

Из данных выражений следует, что материальные точки (небесные тела в приближении материальными точками) движутся по коническим сечениям, а также они устанавливают связь между параметрами конического сечения и постоянными первых интегралов, а следовательно, и начальными условиями движения материальных точек. 

При $\upsilon=0$ величина $r$ имеет наименьшее значение; назовем соответствующую ей точку траектории перицентром, наиболее удаленную точку, достигаемую при $\upsilon=\pi$, – апоцентром. При $\upsilon=0$ вектор $\vec{r}$ направлен вдоль вектора $\vec{f}$, поэтому можно говорить, что вектор Лапласа направлен на перицентр. Прямая, совпадающая с большой полуосью орбиты и направленная вдоль вектора Лапласа, называется линией апсид. 

Как следует из получившихся выражений, эксцентриситет определяется модулем вектора Лапласа. При $0\le e<1$ движение происходит по эллипсу; при $e=1$ – по параболе и при $e>1$ – по гиперболе. 
_Комментарий: на самом деле, если эксцентриситет равен единице, а угловой момент равен нулю, траектория является радиальной, но если есть угловой момент, то траектория представляет собой параболу._

Пользуясь введенной полярной системой координат, найдем первый интеграл, дающий явную связь координат со временем (фактически, это второй интеграл, так как отсутствует зависимость от скорости). Как было показано выше, интеграл площадей в этой системе имеет вид $c = r^2\frac{d\upsilon}{dt}$, так как вектор скорости в цилиндрической системе координат ($r$, $\upsilon$, $z$):

$$\vec{V} = \begin{bmatrix} V_r \\ V_\upsilon \\ V_z \end{bmatrix} = V_r\vec{e_r} + V_\upsilon\vec{e_\upsilon} + V_z\vec{e_z},\ где \ V_r = \dot{r}, \ V_\upsilon=r\dot{\upsilon}, \ V_z=\dot{z}=0.$$

Аналогично 

$$\vec{r}=r\vec{e_r}, \ то \ есть \ \vec{r}=\begin{bmatrix} r \\ 0 \\ 0 \end{bmatrix}.$$

Тогда

$$\vec{r}\times\vec{V} = \det \begin{bmatrix} \vec{e_r} & \vec{e_\upsilon} & \vec{e_z} \\ r & 0 & 0 \\ \dot{r} & r\dot{\upsilon} & 0 \end{bmatrix} = \vec{e_z}(r^2\dot{\upsilon}),$$

$$\vec{r}\times\vec{V}=\vec{c}=c\vec{e_z} \implies c=r^2\dot{\upsilon}=r^2\frac{d\upsilon}{dt}.$$

Подставлем сюда выражение для $r$:

$$c = \frac{p^2}{(1+e \cos(\upsilon))^2}\frac{d\upsilon}{dt}, \ \ 1 = \frac{p^2}{c} \frac{d\upsilon}{dt} \frac{1}{(1+e \cos(\upsilon))^2}.$$

После интегрирования получаем $\left[\int_{a}^{b}f(x)dx = F(b) - F(a) \right]$:

$$\large t-\tau=\frac{p^2}{c}\int_{0}^{\upsilon}\frac{d\upsilon}{\left(1+e \cos\left(\upsilon\right)\right)^2},$$

где $\tau$ – постоянная, именуемая **эпохой** – это момент первого пересечения перицентра движущимся небесным телом. Заметим, что, хотя квадратура формально дает решение задачи, воспользоваться ею при аналитическом изучении движения затруднительно.

Во многих случаях в небесной механике предпочтительна сферическая система координат. Представим себе сферу единичного радиуса, центр которой совпадает с центром масс Земли.

Взяв плоскость $Oxy$, совпадающей с плоскостью земного экватора, рассмотрим дуги больших кругов, получаемые на единичной сфере от сечения ее плоскостью $Oxy$ и плоскостью, в которой происходит движение небесного тела. 

Пусть одна из точек пересечения этих двух кругов лежит в первой четверти плоскости $Oxy$ и пусть этой точке соответствует переход небесного тела из южного полушария в северное. Назовем ее **восходящим узлом орбиты**. Угол между осью $Ox$ и прямой, соединяющей точку $O$ с восходящим узлом, назовем **долготой восходящего узла** и обозначим $\Omega$. Иногда удобно рассматривать сумму двух последних углов $u=\omega+\upsilon$, носящую название **аргумент широты**.

Углы $\Omega$, $i$ полностью определяют положение плоскости орбиты в инерциальном пространстве, а угол $u$ дает направление на мгновенное положение небесного тела. Если в этом направлении отложить величину радиус-вектора небесного тела, то его положение будет полностью определено.

Если прямая $OM$ пересекает единичную сферу в точке $M^\prime$, то очевидно, что $x=r\cos\left(M^\prime,x^\prime \right)$, $y=r \cos\left(M^\prime,y^\prime\right)$, $z=r \cos\left(M^\prime,z^\prime\right)$.

Согласно теореме косинусов сферической тригонометрии:

$$\cos{\left(a\right)}=\cos{\left(b\right)}\cos{\left(c\right)}+\sin{\left(b\right)}\sin{\left(c\right)}\cos{\left(A\right)},$$

$$\cos{\left(A\right)}=-\cos{\left(B\right)}\cos{\left(C\right)}+\sin{\left(B\right)}\sin{\left(C\right)}\cos{\left(a\right)},$$

где строчными буквами даны стороны сферических треугольников (дуги больших кругов, измеряемые соответствующими центральными углами), а прописными – двугранные углы между плоскостями больших кругов, причем противолежащие стороны и углы обозначены одинаковыми буквами.

Применим теорему косинусов к сферическим треугольникам $x^\prime M^\prime\Omega^\prime$, $\Omega^\prime M^\prime y^\prime$ и $z^\prime M^\prime \Omega^\prime$. 

Рассмотрим $x^\prime M^\prime\Omega^\prime$:

$$x=r\cos(M^\prime x^\prime)=r\left[\cos\Omega\cos(\Omega^\prime M^\prime)+\sin\Omega\sin(\Omega^\prime M^\prime)\cos(x^\prime\Omega^\prime M^\prime)\right], \ \Omega^\prime M^\prime=u, \ x^\prime\Omega^\prime M^\prime=180-i, \ \cos(180-\alpha)=-\cos\alpha.$$

Рассмотрим $\Omega^\prime M^\prime y^\prime$:

$$y=r\cos(M^\prime y^\prime)=r\left[\cos(\Omega^\prime y^\prime)\cos(\Omega^\prime M^\prime)+\sin(\Omega^\prime y^\prime)\sin(\Omega^\prime M^\prime)\cos(M^\prime\Omega^\prime y^\prime)\right], \ \Omega^\prime y^\prime=90-\Omega, \ \sin(90-\alpha)=\cos\alpha, \ \cos(90-\alpha)=\sin\alpha.$$

Рассмотрим $z^\prime M^\prime \Omega^\prime$:

$$z=r\cos(M^\prime z^\prime)=r\left[\cos(M^\prime\Omega^\prime)\cos(z^\prime\Omega^\prime)+\sin(M^\prime\Omega^\prime)\sin(z^\prime\Omega^\prime)\cos(z^\prime\Omega^\prime M^\prime)\right], \ \cos(z^\prime\Omega^\prime)=0, \ \sin(z^\prime\Omega^\prime)=1.$$

Таким образом, получаем:

$$x=r\left[\cos{\left(\Omega\right)}\cos{\left(u\right)}-\sin{\left(\Omega\right)}\sin{\left(u\right)}\cos{\left(i\right)}\right],$$

$$y=r\left[\sin{\left(\Omega\right)}\cos{\left(u\right)}+\cos{\left(\Omega\right)}\sin{\left(u\right)}\cos{\left(i\right)}\right],$$

$$z=r \sin\left(u\right)\sin{\left(i\right)}.$$

Если дополнить эти равенства полученными ранее $r=\frac{p}{1+e \cos(\upsilon)}$, $u=\omega+\upsilon$ и $t-\tau=\frac{p^2}{c}\int_{0}^{\upsilon}\frac{d\upsilon}{\left(1+e \cos\left(\upsilon\right)\right)^2}$, то оказывается возможным представить декартовы координаты небесного тела как явные функции времени. 

Что касается величин $\Omega$, $i$, $\omega$, $p$, $e$, $\tau$, то они являются константами первых интегралов движения. Совокупность же получившихся равенств является общим решением системы уравнений $(3)$:

$$\large \frac{d^2\vec{r}}{dt^2}=-\mu\frac{\vec{r}}{r^3}.$$

Таким образом, движение спутника полностью определяется положением плоскости его орбиты в пространстве (то есть положением этой плоскости относительно выбранной системы координат); формой и размерами орбиты; положением орбиты в плоскости движения; моментом прохождения спутника через его перицентр (или через какую-либо другую, вполне определенную точку орбиты).

Вышеуказанный набор параметров, задающих размеры и форму орбиты небесного тела, расположение орбиты в пространстве и место расположения небесного тела на орбите в задаче двух тел традиционно называют **кеплеровыми элементами орбиты** или **оскулирующими элементами орбиты** (оскулирующая орбита — орбита, которую объект имел бы при отсутствии в дальнейшем каких-либо возмущений).

На рисунке выше плоскость орбиты (синяя) пересекает базовую плоскость отсчета (серая). Для спутников, вращающихся вокруг Земли, базовой плоскостью обычно является экваториальная плоскость Земли, а для спутников на солнечных орбитах - плоскость эклиптики. Пересечение плоскостей называется линией узлов, так как оно соединяет центр масс с восходящим ($☊$) и нисходящим узлами. Плоскость отсчета вместе с исходной точкой ($♈︎$ - точка весеннего раводенствия) устанавливают систему отсчета. 

Итак, повторим кеплеровы элементы орбиты. Два элемента определяют форму и размер эллипса:
- **эксцентриситет** ($е$) — параметр, характеризующий форму орбиты, которую можно представить одним из конических сечений (круг, эллипс, парабола, гипербола). По сути, описывает, насколько форма "удлинена" по сравнению с кругом; 
- **фокальный параметр орбиты** ($р$), чей геометрический смысл заключается в том, что он равен половине длины фокальной хорды. Часто параметр заменяют на **большую полуось** ($а$) — сумму расстояний периапсиса и апоапсиса, деленную на два (для классических орбит двух тел большая полуось — это расстояние между центрами тел, а не расстояние между телами от центра масс).

_Комментарий: часто можно увидеть слова "перицентр" и "апоцентр", хотя в иностранной технической литературе рекомендуется использовать периапсис / апоапсис. В общих ситуациях термины "перицентр" и "апоцентр" используются для обозначения крайних точек орбит; "периапсис" и "апоапсис" являются эквивалентными альтернативами, но эти термины также часто относятся к расстояниям, то есть наименьшему и наибольшему расстояниям между небесным телом и его "телом-хозяином" (см. рисунок ниже)._

Следующие два элемента определяют ориентацию плоскости орбиты, в которую вложен эллипс:
- **наклонение** ($i$) — вертикальный наклон эллипса по отношению к плоскости отсчета, измеренный в восходящем узле (где орбита проходит вверх через плоскость отсчета, желтый угол $i$ на рисунке). Для искусственных спутников Земли за плоскость отсчёта обычно выбирают плоскость экватора Земли. Угол наклона измеряется перпендикулярно линии пересечения орбитальной плоскости и базовой плоскости;
- **долгота восходящего узла** ($\Omega$) — определяет угол в базовой плоскости, образуемый между базовым направлением на нулевую точку и направлением на точку восходящего узла орбиты. Нулевая точка — Первая точка Овна ($♈︎$ — точка весеннего равноденствия). Угол измеряется от направления на нулевую точку против часовой стрелки. Он измеряется в базовой плоскости и показан на рисунке как зеленый угол $\Omega$. У орбит с наклонением $i$ равным нулю $\Omega$ не определяется (она, как и наклонение $i$, равна нулю).

_Комментарий: точки равноденствия — точки пересечения небесного экватора и эклиптики (точки весеннего и осеннего равноденствия). В точке весеннего равноденствия Солнце пересекает небесный экватор, переходя из южного полушария небесной сферы в северное._

Оставшиеся два элемента следующие:

- **аргумент перицентра** ($\omega$) определяет ориентацию эллипса в плоскости орбиты как угол, измеряемый от восходящего узла к перицентру (ближайшая точка, в которой объект-спутник подходит к основному объекту, вокруг которого он вращается).

- иногда **эпоху** считают «седьмым» параметром орбиты, а не частью системы отсчета, и шестым параметров называют **истинную аномалию** ($\upsilon$), которая определяет положение орбитального тела вдоль эллипса в определенное время.
